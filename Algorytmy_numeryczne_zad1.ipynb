```python
import math
import sys
import matplotlib.pyplot as plt
import numpy as np
sys.setrecursionlimit(10000000)
from decimal import *

getcontext().prec = 100
tabBledow = []

def silnia(n):
    silnia_wynik = 1
    if (n == 0 | n == 1):
        return 1
    else:
        for i in range(2, n + 1):
            silnia_wynik = silnia_wynik * i
        return silnia_wynik


# ---------------------------------------------
def countCosinus(x, n):
    result = 1.0
    a = 1.0
    for i in range(1, n):
        y = silnia(2 * (i))
        a = a * -1.0
        b = 2 * i
        result += a * ((x ** b) / y)
    return (result)


# ----------------------------------------------
def countCosodKonca(x, n):
    result2 = 0.0
    a = 1.0
    tab = []
    tab.append(1.0)

    for i in range(1, n):
        a = a * -1.0
        b = 2 * i
        y = silnia(2 * i)
        result2 = a * ((x ** b) / y)
        tab.append(result2)
    return sumowanie(tab, n)


# ---------------------------------------------
def sumowanie(arr, n):
    suma = 0.0

    for i in range(n - 1, -1, -1):
        suma += arr[i]

    return suma


# -----------------------------------------------

def cosinusPoczPoprzedni(x, n):
    cosx = 1.0
    silnia = 1.0
    a = 1.0
    last = 1.0
    licz = x * x

    for _ in range(1, n):
        temp_mian = silnia * (silnia + 1.0)
        silnia += 2.0
        a *= -1.0
        cosx += a * last * licz / temp_mian
        last = last / temp_mian * licz

    return cosx


# -----------------------------------------------
def cosinusKoniecPoprzedni(x, n):
    silnia = 1.0
    a = 1.0
    last = 1.0
    licz = x * x

    cosx_values = []
    cosx_values.append(1.0)

    for _ in range(1, n):
        temp_mian = silnia * (silnia + 1.0)
        silnia += 2.0
        a *= -1.0
        cosx_values.append(a * last * licz / temp_mian)
        last = last / temp_mian * licz

    return sumowanie(cosx_values, n)

#--------------------------------------------------
def bladBezwzgledny2(x, wartoscFunkcji):
    global tabBledow
    result = np.cos(x) - wartoscFunkcji
#     result = math.fabs(result)
    tabBledow.append(np.abs(result))
    
#--------------------------------------------------

print("-----------------------")
j = 0
for i in range(-50,50):
    bladBezwzgledny2(i,countCosinus(i,30))
    print(tabBledow[j])
    j = j + 1
print("-----------------------")
print("-----------------------")
print("-----------------------")
tabBledow.clear()
# j = 0
# for i in range(-50,50):
#     bladBezwzgledny2(i,countCosinus(i,80))
#     print(tabBledow[j])
#     j = j + 1
# print("-----------------------")
# print("-----------------------")
# print("-----------------------")
# tabBledow.clear()
# j = 0
# for i in range(-50,50):
#     bladBezwzgledny2(i,countCosinus(i,50))
#     print(tabBledow[j])
#     j = j + 1
# print("-----------------------")
# print("-----------------------")
# print("-----------------------")
# tabBledow.clear()
# j = 0
# for i in range(-500,500):
#     bladBezwzgledny2(i,countCosodKonca(i,50))
#     print(tabBledow[j])
#     j = j + 1
# # print(tabBledow)
# print("-----------------------")
# print("-----------------------")
# print("-----------------------")
# tabBledow.clear()
# j = 0
# for i in range(-50,50):
#     bladBezwzgledny2(i,cosinusPoczPoprzedni(i,50))
#     print(tabBledow[j])
#     j = j + 1
# # print(tabBledow)
# print("-----------------------")
# print("-----------------------")
# print("-----------------------")
# tabBledow.clear()
# j = 0
# for i in range(-5000,5000):
#     bladBezwzgledny2(i,cosinusKoniecPoprzedni(i,50))
#     print(tabBledow[j])
#     j = j + 1
# # print(tabBledow)
# print("-----------------------")
# print("-----------------------")
# print("-----------------------")

```
